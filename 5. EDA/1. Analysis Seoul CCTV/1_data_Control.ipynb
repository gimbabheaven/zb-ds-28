loc : location
- index 이름으로 특정 행, 열을 선택
- df.loc[행, 열] : 이름으로도 사용 가능, pandas의 보편적 옵션

```python
df.loc[:, ["A", "B"]] # 인덱스는 전체, 컬럼은 A, B
```

```python
df.loc["2024-07-06":"2024-07-10", ["A", "D"]]
```

```python
df.loc["2024-07-06":"2024-07-10", "A":"D"]
```

```python
df.loc["2024-07-06", ["A", "D"]]
```

iloc : integer location
- 컴퓨터가 인식하는 인덱스 값으로 선택
- df.iloc[행 번호, 열 번호] : 열 번호로 접근하여 슬라이싱

```python
df.iloc[3]
```

```python
df.iloc[3, 2] # 좌표
```

```python
df.iloc[3:5, 0:2] # 3, 4행 0, 1열 예상
```

```python
df.iloc[[1, 2, 4], [0,2]]
```

```python
df.iloc[:, 1:3]
```
