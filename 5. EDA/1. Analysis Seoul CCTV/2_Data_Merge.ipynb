## 데이터 합치기

### Pandas에서 데이터 프레임을 병합하는 방법
- pd.concat()
- pd.merge()
- pd.join()

```python
# 딕셔너리 안의 리스트 형태
left = pd.DataFrame({
    "Key" : ["K0", "K4", "K2", "K3"],
    "A" : ["A0", "A1", "A2", "A3"],
    "B" : ["B0", "B1", "B2", "B3"]
})
left
```

```python
# 리스트 안의 딕셔너리 형태
right = pd.DataFrame([
    {"Key":"K0", "C":"C0", "D":"D0"},
    {"Key":"K1", "C":"C1", "D":"D1"},
    {"Key":"K2", "C":"C2", "D":"D2"},
    {"Key":"K3", "C":"C3", "D":"D3"},
])
right
```

#### pd.merge()
- 두 데이터 프레임에서 컬럼이나 인덱스를 기준으로 잡고 병합하는 방법
- 기준이 되는 컬럼이나 인덱스를 "Key"값
- 기준이 되는 "Key"값은 두 데이터 프레임에 모두 포함되어 있어야 함

```python
# "Key"라는 이름의 컬럼을 기준으로 병합
pd.merge(left, right, on="Key") # how 디퐅트 = inner
```

```python
# "Key"라는 이름의 컬럼을 기준으로 left join
pd.merge(left, right, how="left", on="Key") # 공백 데이터는 NaN
```

```python
# "Key"라는 이름의 컬럼을 기준으로 right join
pd.merge(left, right, how="right", on="Key") # 공백 데이터는 NaN
```

```python
# "Key"라는 이름의 컬럼을 기준으로 full outer join
pd.merge(left, right, how="outer", on="Key") # 공백 데이터는 NaN
```
