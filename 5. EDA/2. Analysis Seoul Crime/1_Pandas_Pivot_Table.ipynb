```python
import pandas as pd
import numpy as np
```

## Pandas Pivot Table

- index, columns, values, aggfunc

```python
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/DS/rawdata/EDA/02. sales-funnel.xlsx')
df.head()
```

### index 설정

```python
# Name 컬럼을 인덱스로 설정
# pd.pivot_table(df, index="Name") > Type Error: could not convert string
pd.pivot_table(df, index='Name', values=['Account', 'Price', 'Quantity'], aggfunc='sum')
```

```python
## multi index 설정
df.pivot_table(index=['Name', 'Rep', 'Manager'], values=['Account', 'Price', 'Quantity'], aggfunc='sum')
```

```python
df.pivot_table(index=['Manager', 'Rep'], values=['Account', 'Price', 'Quantity'], aggfunc='sum')
```

### values 설정

```python
df.pivot_table(index=['Manager', 'Rep'], values='Price', aggfunc='mean')
```

```python
# Price 컬럼에 Sum 연산 적용
df.pivot_table(index=['Manager', 'Rep'], values='Price', aggfunc=np.sum)
```

```python
df.pivot_table(index=['Manager', 'Rep'], values='Price', aggfunc=[np.sum, len])
```

### columns 설정

```python
# Product를 컬럼으로 지정
df.pivot_table(index=['Manager', 'Rep'], values='Price', columns='Product', aggfunc=np.sum)
```

```python
# Nan 값 설정 : fill_value
df.pivot_table(index=['Manager', 'Rep'], values='Price', columns='Product', aggfunc=np.sum, fill_value=0)
```

```python
# 2개 이상의 index, values 설정
df.pivot_table(index=['Manager', 'Rep', 'Product'], values=['Price', 'Quantity'], aggfunc=np.sum, fill_value=0)
```

```python
# aggfunc 2개 이상 설정
df.pivot_table(
    index=['Manager', 'Rep', 'Product'],
    values=['Price', 'Quantity'],
    aggfunc=[np.sum, np.mean],
    fill_value=0,
    margins=True) # 총계 (All) 추가
```
